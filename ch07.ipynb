## matplotlib Architecture
- must provide a set of functions and tools that allow representation and manipulation of a _Figure_ (main object), along with all internal objects of which it is composed
- not only deals with graphics but animations too (key or click)
    - communication in the architecture is unidirectional (lower lvls can't talk to top lvls)
        - top: scripting      ↓
        - middle: artist      ↓
        - bottom: backend  ↓

## Backend Layer
- lowest level of matplotlib which contains the APIs which are as et of classes that play a role of implementation of graphic elements at a low level
    - FigureCanvas is object that embodies a drawing area
    - Renderer is object that draws on FigureCanvas
    - Event is the object that handles user inputs (key and mouse events)

## Artist Layer
 - intermediate level
 - all elements that make up the chart such as a title, axis labels, markers, etc., are instances of the Artist object with 2 main classes:
     - _primitive artists_: individual objects that constitute the basic elements to form a graphical representation in a plot, for ex. a Line2D or as a geometric figure such as a Rectangle or Circle and even pieces of text
     - _composite artists_: graphic elements present in a chart that are composed of several basic elements: namely the primitive artists. examples include axis, ticks, axes, and figures
         -_figure_ is the highest lvl on the hierarchy
         -_axes_: generally what is eant as plot or chart
         - _axis_: manages ticks etc.

## Working w/ Multiple Figures and Axes
- allows you to create subplots -> always keep in mind current figure and current axes(plot within figure)
 - arg set to subplot sets mode of subdivision
     - first number gives vertical divisions
     - second number gives horizontal divisions
     - third number selects which is current subplot on which you can direct commands

```python
t = np.arange(0,5,0.1)
y1 = np.sin(2*np.pi*t)
y2 = np.sin(2*np.pi*t)
plt.subplot(211) #2 vert, 1 hor, 1st is current subplot
plt.plot(t,y1,'b-.')
plt.subplot(212) #2 vert, 1 hor, 2nd is current subplot
plt.plot(t,y2,'r--')
```

```python
t = np.arange(0.,1.,0.05)
y1 = np.sin(2*np.pi*t)
y2 = np.cos(2*np.pi*t)
plt.subplot(121) #1 vert, 2 hor, 1st is current subplot
plt.plot(t,y1,'b-.')
plt.subplot(122) #1 vert, 2 hor, 2nd is current subplot
plt.plot(t,y2,'r--')
```

```python
x = np.arange(-2*np.pi, 2*np.pi, 0.01)
y = np.sin(x)/x
y2 = np.sin(2*x)/x
y3 = np.sin(3*x)/x
plt.plot(x,y,color='b')
plt.plot(x,y2,color='r')
plt.plot(x,y3,color='g')
plt.xticks([-2*np.pi,-np.pi,0,np.pi,2*np.pi],
          [r'$-2\pi$',r'$-\pi$',r'$0$',r'$+\pi$',r'$+2\pi$'])
plt.yticks([-1,0,1,2,3],
          [r'$-1$',r'$0$',r'$+1$',r'$+2$',r'$+3$'])
plt.annotate(r'$\lim_{x\to 0}\frac{\sin(x)}{x} = 1$',xy=[0,1],
            xycoords='data',xytext=[30,30],fontsize=16,textcoords='offset points',
            arrowprops=dict(arrowstyle="->",connectionstyle="arc3,rad=.2"))
ax = plt.gca() #captures Axes object, then thru this we can define the bounding box
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
ax.yaxis.set_ticks_position('left')
ax.spines['bottom'].set_position(('data',0))
```

- annotate() function is very good for things like the graph above, of course we see the LateX and fontsize kwargs, the point of the chart indicated by a list[x,y] , distance of the text is defined by xytext,and curved arrow's characteristics is set by arrowprops(made curvy by the arc and rad)

- can add std. dev. through yerr kwarg along with a list containing the std dev
- usually combined with error_kw, represents error bars
- eColor/ capsize - width of transverse lines that mark error bars
- alpha indicates transparency (0(transparent)-1(solid))

```python
index = np.arange(5)
values1 = [5,7,3,4,6]
std1 = [0.8,1.0,0.4,0.9,1.3]
plt.title('A Bar Chart')
plt.bar(index, values1,yerr=std1,error_kw={'ecolor':'0.1','capsize':6},alpha=0.7,
       label ='First')
plt.xticks(index+0.4,['A','B','C','D','E'])
plt.legend(loc=2)
```

## Contour Plots
- suitable for displaying 3D surfaces through a contour map
- implementation is not bad: need function $z=f(x,y)$ for the 3D surface, where x and y will define area of map
    - z values come from ordered pairs (a mtrix of values)

```python
dx = 0.01; dy= 0.01
x = np.arange(-2.0,2.0,dx) #obviously dx is increment
y = np.arange(-2.0,2.0,dy)
X,Y = np.meshgrid(x,y)
def f(x,y):
    return (1 - y**5 + x**5)*np.exp(-x**2-y**2)
C = plt.contour(X,Y,f(X,Y),8,colors='black')
plt.contourf(X,Y,f(X,Y),8)
plt.clabel(C, inline=1,fontsize=10)
```
